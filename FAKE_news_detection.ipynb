```python
def wordopt(text):
  text = text.lower()
  text = re.sub('\[.*?]','',text)
  text = re.sub("\\W"," ",text)
  text = re.sub('http?://\S+|www\.\S+','',text)
  text = re.sub('<.*?>+','',text)
  text = re.sub('[%s]'%re.escape(string.punctuation),'',text)
```
